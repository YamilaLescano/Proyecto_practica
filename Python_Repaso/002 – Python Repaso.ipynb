### Ejercicio 3 – Listas y bucles

```
lista = [2,6,8,9,1]

for numero in lista :
    print(numero)
```

```
lista = [2,6,8,9,1]
suma = 0
count = 0

for numero in lista:
    suma = suma + numero
    count =count + 1
print("La suma de los numeros es :" , suma)
print("El promedio de los numeros es :" , suma/count)
```
