### Ejercicio 2 – Dataset simple con Pandas

```
data = {
    "estudiante": ["Ana", "Juan", "Carla", "Pedro", "Lucia"],
    "nota": [8, 6, 9, 5, 7]
}

df = pd.DataFrame(data)

print(df)
```

```
promedio = df["nota"].mean()
maxima = df["nota"].max()
minima = df["nota"].min()

print("Promedio:", promedio)
print("Nota máxima:", maxima)
print("Nota mínima:", minima)
```
